```python
x=df.drop("species",axis=1)      #independent variable
x
```

```python
y=df.species  #dependent variable
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors=4)
```

```python
model.fit(train_x,train_y)
```

```python
model.score(test_x,test_y)
```

```python
model.predict([[1.9,2.3,4.5,8.9]])
```

```python
import warnings 
warnings.filterwarnings("ignore")
```

```python
sl=float(input("sepal length : "))
sw=float(input("sepal width : "))
pl=float(input("petal length : "))
pw=float(input("petal width : "))
if model.predict([[sl,sw,pl,pw]])==0:
    print("setosa")
if model.predict([[sl,sw,pl,pw]])==1:
    print("versicolor")
else:
    print("virginica")
```

hit and trail method to find the best value of k

```python
for i in range(1,50):
    model=KNeighborsClassifier(n_neighbors=i)
    model.fit(train_x,train_y)
    model.score(test_x,test_y)
    print(f"value of K is : {i} then score is {model.score(test_x,test_y)}")
```
